```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
```

```python
dt= 1/24
x = np.arange(0,50)
t = np.arange(0,10+dt,dt)
rad = 2*np.pi
omega = rad*0.1
L = 100
u = 1
```

```python
ds = xr.Dataset({'x':(['x'],x),'t':(['t'],t)})
```

```python
ds['fluc'] = np.exp(1j*(omega*ds.t-rad/L*ds.x))
```

```python
ds['fluc'].real.plot()
```

```python
ds.sel(x=25,method='nearest').fluc.imag.plot()
```

```python
ds.fluc.differentiate('t')
```
