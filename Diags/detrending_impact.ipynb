```python
path = '/Users/zcaspar/codes/Diagnostics/'
_uu = xr.open_dataset(path+'uucorr_20.nc')
_vv = xr.open_dataset(path+'vvcorr_20.nc')
_uv = xr.open_dataset(path+'uvcorr_20.nc')
_uuEL = xr.open_dataset(path+'diags/corr_20.nc')
uuE = _uuEL.E_corr
```

```python
# Energy : 
Estats = xr.open_dataset('/Users/zcaspar/codes/Diagnostics/diags/statsE.nc')
Lstats = xr.open_dataset('/Users/zcaspar/codes/Diagnostics/diags/statsL.nc')

KE_l = 0.5*np.sqrt(Lstats.u_mean.mean('x_bins')**2+Lstats.v_mean.mean('x_bins')**2)
```

```python
utrend = _uv.trend_var_0
vtrend = _uv.trend_var_1
KEtrend = 0.5*np.sqrt(utrend**2+vtrend**2)
```

```python
fig, ax = plt.subplots(1,2,figsize=(15,7))
KE_l.plot(ax=ax[0],y='y_bins',label=r'KE = 0.5$\sqrt{\bar u^2+ \bar v^2}$')
Lstats.u_mean.mean('x_bins').plot(ax=ax[0],y='y_bins',label=r'$\bar u$')
Lstats.v_mean.mean('x_bins').plot(ax=ax[0],y='y_bins',label=r'$\bar v$')
ax[0].legend()
KEtrend.sel(lag=0).plot(ax=ax[1],y='y_bins')
utrend.sel(lag=0).plot(ax=ax[1],y='y_bins')
vtrend.sel(lag=0).plot(ax=ax[1],y='y_bins')
```
